```python
# Defining the base URL of REDFIN site to select data from California
url = "https://www.redfin.com/state/California"

req = requests.Session().get(url, headers=req_headers)

redfin_soup = BS(req.content, 'html.parser')
```

```python
# Identifying all the hrefs and getting all the 
# possible redirection links from above defined REDFIN site

cal_urls = [a['href'] for a in redfin_soup('a')  if a.has_attr('href')]

cal_urls
```

```python
# Defining REGULAR EXPRESSION to extarct all the CALIFORNIA City URLS for targeted data scrapping

city_regex = r"https://www.redfin.com/city/?"

city_urls = [url for url in cal_urls if re.match(city_regex, url)]

city_urls1 = list(set(city_urls))

city_urls1
```

```python
# Defining REGULAR EXPRESSION to extarct any other possible page links to california citiesin redfin
city_regex2 = r"/city/?"

city_urls2 = [url for url in cal_urls if re.match(city_regex2, url)]

city_urls2 = list(set(city_urls2))

city_urls2
```

```python
# Defining an empty list to store the newly generated redfin urls for California cities

city_urls3 = []

for url in city_urls2:
    
    new_urls = "https://www.redfin.com" + url
    
    city_urls3.append(new_urls)

city_urls3
```

```python
# Creating a new list to save all the possible urls of house listings in CALIFORNIA at CITY level 

new_city_urls = city_urls1 + city_urls3

# Prining the URLs from which we will be scraping data

new_city_urls
```

```python
# Filtering the URLs based on cities in BAY AREA

# Crwating new empty list to store the filteres URLs.
city_list = []

cities = ["San-Francisco","San-Jose","Pleasanton",
          "Fremont","San-Ramon","Oakland","Dublin"]

for url in new_city_urls:
    
    if url.split("/")[6] in cities:
        
        city_list.append(url)
city_list
```

```python
# Running a loop to extarct individual house listing level URL info from all the possible pages

BayAreaListings = []

for i in city_list:
    BayAreaListings.append(i)
    for j in range(2,15):
        BayAreaListings.append(i+'/page-'+str(j))
print(BayAreaListings)

houseListing = defaultdict(set)

for url in BayAreaListings:
    city_name = re.findall('CA/([^/]+)',url) # key
    text = requests.get(url, headers=req_headers).text
    house_soup = BS(text,'html5lib')
    city_text = house_soup.find('div',attrs={'id':'results-display'})
    house_links = [a['href'] for a in city_text('a') if a.has_attr('href')]
    ca_regex = r"^/CA"
    for link in house_links:
        if re.match(ca_regex,link):
            mlink = "https://www.redfin.com" + link
            houseListing[city_name[0]].add(mlink)

#print(len(houseListing))
print(houseListing)
```

```python
# Creating empty lists for required parameters

Address_Line= []
State_Code = []
City = []
Zipcode = []
Property_Price = []
Bed_Rooms = []
Bath_Rooms = []
Prop_Sqft = []
School_score = []
Garage = []
Time_on_redfin = []
Property_type = []
Year_Built = []
Style = []
Community =[]

# Defining a function for Regex to scrape few additional varibales

def find_between_strings(input_string, start_string, end_string):
    pattern = re.escape(start_string) + "(.*?)" + re.escape(end_string)
    matches = re.finditer(pattern, input_string)
    results = [match.group(1) for match in matches]
    return results



for city in houseListing:


    for ba_prop in houseListing[city]:

        City.append(city)

        html_text = requests.get(ba_prop,headers=req_headers).text
        ba_soup = BS(html_text,'html5lib')
        
        try:
            address = ba_soup.find('div',attrs = {'class':'street-address'}).text.split(',')[0]
        except:
            address = np.nan
        Address_Line.append(address)        
        
        try:
            state_code = ba_soup.find('div',attrs = {'class':'dp-subtext'}).text.split(', ')[1].split(' ')[0]
        except:
            state_code = np.nan
        State_Code.append(state_code)

        try:
            zip_code = ba_soup.find('div',attrs = {'class':'dp-subtext'}).text.split(', ')[-1].split(' ')[-1]

        except:
            zip_code = np.nan
        Zipcode.append(zip_code)


        try:
            prop_price = ba_soup.find('div',attrs = {'class':'statsValue'}).text.split('$')[1]
        except:
            prop_price = np.nan
        Property_Price.append(prop_price)

        try:
            bed_cnt = ba_soup.find('div',attrs = {'data-rf-test-id':'abp-beds'}).text.split('B')[0]
            bed_val = int(bed_cnt)
        except:
            bed_val=np.nan
        Bed_Rooms.append(bed_val)


        try:
            bath_cnt =ba_soup.find('div',attrs = {'data-rf-test-id':'abp-baths'}).text.split('B')[0]
            bath_val = float(bath_cnt)
        except:
            bath_val = np.nan
        Bath_Rooms.append(bath_val)


        try:
            sqft_area = ba_soup.find('div',attrs = {'data-rf-test-id':'abp-sqFt'}).text.split('S')[0].replace(',','')
            sqft_val = float(sqft_area)
        except:
            sqft_val = np.nan
        Prop_Sqft.append(sqft_val)



        try:
            schl_rat = max([rate.text[0] for rate in ba_soup.findAll('div',attrs = {'class':'gs-rating-text'})])
        except:
            schl_rat = np.nan

        School_score.append(schl_rat)


        try:
            garage_aval = ba_soup.findAll('span',attrs = {'class':'entryItemContent'})
            for tag in garage_aval:
                if("Garage" in tag.text):
                    value = 'Yes'
                    break
                else:
                    value = 'No'
        except:
             value = np.nan
        Garage.append(value)
        
        try:
            div_year_built = ba_soup.findAll('div',attrs={'class':'keyDetailsList'})
            input_text=str(div_year_built[0])
            start_marker = '<span class="table-label">Year Built</span><div class="table-value">'
            end_marker = "</div>"
            results = find_between_strings(input_text, start_marker, end_marker)
            if results:
                res=results[0]
            else:
                res=np.nan
           # print("Year Built", res)
        except:
            res=np.nan
        Year_Built.append(res)

        
        try:
            time_on = ba_soup.findAll('div',attrs={'class':'keyDetailsList'})
            input_text=str(time_on)
            start_marker = '<span class="bp-DefinitionFlyout bp-DefinitionFlyout__underline" role="link" tabindex="0">Time on Redfin</span></div></span><div class="table-value">'
            end_marker = "</div>"
            results = find_between_strings(input_text, start_marker, end_marker)
            if results:
                res=results[0]
            else:
                res=np.nan
            #print("time", res)
        except:
            res=np.nan
        Time_on_redfin.append(res)

        
        try:
            prp_type = ba_soup.findAll('div',attrs={'class':'keyDetailsList'})
            input_text=str(prp_type)
            start_marker = '<span class="table-label">Property Type</span><div class="table-value">'
            end_marker = "</div>"
            results = find_between_strings(input_text, start_marker, end_marker)
            if results:
                res=results[0]
            else:
                res=np.nan
           # print("proptype", res)
        except:
            res=np.nan
        Property_type.append(res)

        
        try:
            prp_style = ba_soup.findAll('div',attrs={'class':'keyDetailsList'})
            input_text=str(prp_style)
            start_marker = '<span class="table-label">Style</span><div class="table-value">'
            end_marker = "</div>"
            results = find_between_strings(input_text, start_marker, end_marker)
            if results:
                res=results[0]
            else:
                res=np.nan
            #print("style", res)
        except:
            res=np.nan
        Style.append(res)

        try:
            prp_comm = ba_soup.findAll('div',attrs={'class':'keyDetailsList'})
            input_text=str(prp_comm)
            start_marker = '<span class="table-label">Community</span><div class="table-value">'
            end_marker = "</div>"
            results = find_between_strings(input_text, start_marker, end_marker)
            if results:
                res=results[0]
            else:
                res=np.nan
            #print("communi", res)
        except:
            res=np.nan
        Community.append(res)
```

```python
# Loading teh scraped data into a data frame

ba_house_listings = pd.DataFrame(data = {'AddressLine':Address_Line,
                        'StateCode':State_Code,
                        'City':City,
                        'Zipcode':Zipcode,
                        'PropertyPrice':Property_Price,
                        'BedRooms':Bed_Rooms,
                        'BathRooms':Bath_Rooms,
                        'Property_Sqft':Prop_Sqft,
                        'School_rating':School_score,
                        'Garage':Garage,
                        'YearBuilt':Year_Built,
                        'TimeOnRedfin':Time_on_redfin,
                        'PropertyType':Property_type,
                        'PropertyStyle':Style,
                        'PropertyCommunity':Community})
```

```python
# Saving the raw data(unclean) to a csv file

ba_house_listings.to_csv('BAN_612_Group5_House_Market_Final_unclean_Data.csv')
```

# DATA CLEANSING

```python
# Identifying the vaiables and their data types in the scrapped data

ba_house_listings.info()
```

```python
# Dropping duplicate listings based on the Address Line 

ba_house_listings = ba_house_listings.drop_duplicates(subset = ['AddressLine'])

ba_house_listings
```

```python
# Reindexing

ba_house_listings = ba_house_listings.reindex()
ba_house_listings
```

```python
# Replacing the NULL values in Property Style column with the most occuring style

most_common_style = ba_house_listings['PropertyStyle'].mode()[0]

ba_house_listings['PropertyStyle'].fillna(most_common_style, inplace=True)

ba_house_listings
```

```python
# Defining a function to modify data in Time On Redfin column

def convert_to_days(time_str):
    if isinstance(time_str, str):
        if 'day' in time_str:
            return int(time_str.split()[0])
        elif 'hour' in time_str:
            return float(time_str.split()[0]) / 24
    return np.nan

# Convert "TimeOnRedfin" column to numerical format (days)

ba_house_listings['TimeOnRedfin_Days'] = ba_house_listings['TimeOnRedfin'].apply(convert_to_days)

# Replacing the null values in "TimeOnRedfin" with the average time

average_time_on_redfin = ba_house_listings['TimeOnRedfin_Days'].mean()

ba_house_listings['TimeOnRedfin_Days'].fillna(average_time_on_redfin, inplace=True)


ba_house_listings
```

```python
# Identifying the columns which has NULL values

ba_house_listings.isna().sum()
```

```python
# Dropping rows that has NULL values to data quality issues

ba_house_listings = ba_house_listings.dropna()

ba_house_listings
```

```python
# Replacing the special characters like + and , in the Property Price column

# Remove plus signs '+'
ba_house_listings['PropertyPrice'] = ba_house_listings['PropertyPrice'].str.replace('+', '')

# Replacing ',' using string replace
        
ba_house_listings['PropertyPrice'] = ba_house_listings['PropertyPrice'].str.replace(',','')
```

```python
ba_house_listings['PropertyPrice'] = ba_house_listings['PropertyPrice'].astype(float)
```
